# Transformation des donnÃ©es en Data Frame:

```python
AAPL_2019 = pd.DataFrame(AAPL_2019)
```

# Prediction 50 days into the future

```python
future_days = 50
AAPL_2019['Prediction'] = AAPL_2019['Close'].shift(-future_days)
```

```python
AAPL_2019.head()
```

```python
AAPL_2019.tail()
```

Creation d'un tableau  qui porte les cours sans les Predictions:

```python
X = np.array(AAPL_2019.drop(['Prediction'], 1))[:-future_days]
print(X)
```

```python
#la taille du tableau:
X.shape
```

Creation d'un tableau  qui porte les Predictions sans les cours :

```python
y = np.array(AAPL_2019['Prediction'])[:-future_days]
print(y)
```

```python
#taille du tableau:
y.shape
```

# Split Dataset (Training and Test) :

# Chargements des bibliothÃ©ques Ã  utiliser:

# Sklearn :

=> Sklearn est une bibliotheque d'apprentissage automatique  

Train test plit : x_train , x_test , y_train , y_test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

# linearRegression:

```python
#linearRegression:
from sklearn.linear_model import LinearRegression
```

```python
# Implementing Linear Regression Algorithms.
lr = LinearRegression().fit(x_train, y_train)
```

```python
x_future = AAPL_2019.drop(['Prediction'], 1)[:-future_days]
x_future = x_future.tail(future_days)
x_future = np.array(x_future)
x_future
```

# Prediction:

```python
lr_prediction = lr.predict(x_future)
print(lr_prediction)
```

# Validation

```python
predictions = lr_prediction
valid = AAPL_2019[(X.shape[0]):]
valid['Predictions'] = predictions
```

# Graphique : Original , valide , prediction : 

```python
plt.figure(figsize=(15,10))
plt.title("Apple's Stock Price Prediction Model(Linear Regression Model)")
plt.xlabel('Days')
plt.ylabel('Close Price USD ($)')
plt.plot(AAPL_2019['Close'])
plt.plot(valid[['Close', 'Predictions']])
plt.legend(["Original", "Valid", 'Predicted'])
plt.show()


```
